```python
import numpy as np
from hydro import *
```

```python
#Instance creation with parameters specification
myhdro = hydro(c = np.array([2.,5.]),u_max = [10.,15.],v_max = [25.,30.],ρ = [0.7,0.5])
#Future cost function computation
myfcf = myhdro.compute_fcf(T_max = 12)

```

```python
#Plot fcf matrix as a plane and as a scaled image
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook
%matplotlib notebook
fig = plt.figure()
ax1 = plt.subplot(121,projection='3d')
S0 = myhdro.S0; S1 = myhdro.S1
S0g,S1g = np.meshgrid(S0,S1)
ax1.plot_surface(S0g,S1g,myhdro.fcf[6],cmap = 'hot_r')
ax1.set_xlabel('S0')
ax1.set_ylabel('S1')
ax1.set_zlabel('fcf')
ax2 = plt.subplot(122)
plt.pcolor(S0,S1,np.array(myhdro.fcf[6]),cmap = 'hot_r')
ax2.set_xlabel('S0')
ax2.set_ylabel('S1')
plt.colorbar()
```
